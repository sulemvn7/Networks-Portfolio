```python
import hashlib
import time
```

```python
def brute_force_simulator(target_hash: str, hash_type: str = "md5"):
    """
    Simulate a dictionary attack on a hashed password.
    
    Parameters:
    - target_hash: str, the hash we're trying to crack
    - hash_type: str, "md5" or "sha256"
    """
    
    # Small example dictionary (common passwords)
    common_passwords = ["123456", "password", "123456789", "qwerty", 
                        "abc123", "password1", "Pass123", "MyP@ssw0rd"]
    
    print(f"üîé Starting brute-force simulation using {hash_type.upper()}...")
    start_time = time.time()
    
    for pwd in common_passwords:
        # Hash the password using specified type
        if hash_type.lower() == "md5":
            hashed = hashlib.md5(pwd.encode()).hexdigest()
        elif hash_type.lower() == "sha256":
            hashed = hashlib.sha256(pwd.encode()).hexdigest()
        else:
            raise ValueError("Unsupported hash_type. Use 'md5' or 'sha256'.")
        
        # Compare with target hash
        if hashed == target_hash:
            elapsed = time.time() - start_time
            print(f"‚úÖ Password found: '{pwd}'")
            print(f"‚è± Time taken: {elapsed:.4f} seconds")
            return pwd
    
    elapsed = time.time() - start_time
    print("‚ùå Password not found in dictionary.")
    print(f"‚è± Time taken: {elapsed:.4f} seconds")
    return None
```

```python
# Example: target hash of "Pass123" using MD5
target_md5 = hashlib.md5("Pass123".encode()).hexdigest()
target_sha256 = hashlib.sha256("MyP@ssw0rd".encode()).hexdigest()

# Run the brute force simulator
brute_force_simulator(target_md5, hash_type="md5")
brute_force_simulator(target_sha256, hash_type="sha256")
```
