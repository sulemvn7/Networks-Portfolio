```python
import pyotp
import qrcode

# Generate a new secret key
secret = pyotp.random_base32()
print("Your secret key:", secret)

# Create a TOTP object
totp = pyotp.TOTP(secret)

# Create provisioning URI (used by Google Authenticator / Authy)
uri = totp.provisioning_uri(name="your_email@example.com", issuer_name="My2FADemo")
print("Provisioning URI:", uri)

# Create QR code
img = qrcode.make(uri)
img.save("totp_qr.png")

print("QR code saved as totp_qr.png — scan it with Google Authenticator")
```

```python
import time

print("Your current TOTP codes (changes every 30 seconds):")

while True:
    print("Current OTP:", totp.now())
    time.sleep(5)
```
