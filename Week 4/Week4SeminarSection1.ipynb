```python
import os
import hashlib
import csv
from datetime import datetime

# Path to the folder you created
folder_path = "integrity_files"

# Output CSV file
output_file = "file_hashes.csv"

def compute_sha256(file_path):
    """Compute SHA256 hash for a file."""
    sha256_hash = hashlib.sha256()
    with open(file_path, "rb") as f:
        for block in iter(lambda: f.read(4096), b""):
            sha256_hash.update(block)
    return sha256_hash.hexdigest()

# List to store results
results = []

# Loop through files in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    
    if os.path.isfile(file_path):
        file_hash = compute_sha256(file_path)
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        results.append([filename, file_hash, timestamp])

# Write results to CSV
with open(output_file, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["File Name", "SHA-256 Hash", "Timestamp"])
    writer.writerows(results)

print("File integrity baseline saved to:", output_file)
```
