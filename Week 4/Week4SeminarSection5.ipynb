```python
import psutil
import socket
import time
import logging
from datetime import datetime
```

```python
# Configure logging
logging.basicConfig(
    filename="network_monitor.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

print("‚úÖ Logging system initialized. All anomalies will be recorded in 'network_monitor.log'.")
```

```python
def monitor_network(threshold=50, duration=10):
    print("üîç Monitoring network connections...")
    start_time = time.time()

    while time.time() - start_time < duration:
        connections = psutil.net_connections(kind='inet')
        outbound = [conn for conn in connections if conn.status == 'ESTABLISHED' and conn.raddr]

        if len(outbound) > threshold:
            alert_msg = f"‚ö†Ô∏è ALERT: {len(outbound)} outbound connections detected! Possible anomaly."
            print(alert_msg)
            logging.warning(alert_msg)
        
        time.sleep(1)

    print("‚úÖ Monitoring completed.")
```

```python
def summarize_activity():
    connections = psutil.net_connections(kind='inet')
    active_conns = [conn for conn in connections if conn.status == 'ESTABLISHED' and conn.raddr]
    
    unique_ips = set([conn.raddr.ip for conn in active_conns if conn.raddr])
    summary = f"Summary: {len(active_conns)} active outbound connections to {len(unique_ips)} unique IPs."

    print(summary)
    logging.info(summary)
```

```python
def layered_defence_monitor():
    print("üöÄ Starting Layered Defence Network Monitor")
    logging.info("=== Layered Defence Monitor Started ===")
    
    try:
        # Detection Phase
        monitor_network(threshold=50, duration=15)
        
        # Logging & Reporting Phase
        summarize_activity()

        print("‚úÖ Defence cycle complete. Logs saved in 'network_monitor.log'.")
    except Exception as e:
        logging.error(f"Error occurred: {e}")
        print("‚ùå Error:", e)
```

```python
layered_defence_monitor()
```
